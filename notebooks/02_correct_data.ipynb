#### ü•≠ **`application_train.csv`**

```python
df_train = pd.read_csv("../data_raw/application_train.csv")
df_test  = pd.read_csv("../data_raw/application_test.csv")

# Comparar columnas y devolver las diferencias
columns_train = list(df_train.columns)
columns_test = list(df_test.columns)

# Identificar columnas que est√°n en df_train pero no en df_test
diff_train = [col for col in columns_train if col not in columns_test]

# Identificar columnas que est√°n en df_test pero no en df_train
diff_test = [col for col in columns_test if col not in columns_train]

# Mostrar las diferencias
if diff_train or diff_test:
    print("Diferencias encontradas:")
    if diff_train:
        print(f"Columnas en df_train pero no en df_test: {diff_train}")
    if diff_test:
        print(f"Columnas en df_test pero no en df_train: {diff_test}")
else:
    print("Ambos DataFrames tienen las mismas columnas.")
```
